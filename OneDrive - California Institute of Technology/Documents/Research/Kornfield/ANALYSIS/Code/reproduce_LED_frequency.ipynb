This file reproduces the time-series plot of the intensity of Prof. Julie Kornfield's LED bike light as measured by a Chronos high-speed camera. This method back-calculates the frequency of the LED.

```python
import numpy as np
import matplotlib.pyplot as plt

# user parameters
sampling_freq = 2111 # Hz
light_freq = 999.2 # Hz
shift = 5*np.pi/4
t_i = 0.095 # seconds
t_f = 0.149 # seconds
mean_intensity = 73
amplitude = 3
# saving
savePlot = False
saveFolder = '..\\High-speed Camera\\mean_intensity\\'
```

```python
# derived parameters
nPts = (t_f-t_i)*sampling_freq # number of points to sample
t = np.linspace(t_i, t_f, nPts+1) # +1 to account for endpoints
```

```python
# plots
light_intensity = mean_intensity + amplitude*np.sin(2*np.pi*light_freq*t+shift)
plt.plot(t, light_intensity, label=str(light_freq) + ' Hz')
plt.xlabel('time (seconds)', fontsize=16)
plt.ylabel('intensity (a.u.)', fontsize=16)
plt.title('Reproduction of Intensity vs. Time for High-speed Camera')
plt.legend()
if savePlot:
    saveName = 'intensity_vs_time_' + str(light_freq) + 'Hz.png'
    plt.savefig(saveFolder + saveName, bbox_inches='tight')
plt.show()
plt.close()
```
