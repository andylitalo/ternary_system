```python
import numpy as np
import matplotlib.pyplot as plt

# user parameters
ax_fs = 16 # axis labels
t_fs = 18 # title
tk_fs = 14 # tick marks
l_fs = 14 # legend
```

## VORANOL 360

T = 25 and 60 C

```python
# data from 20190220_co2_solubility_v360_2110b.csv

# pressure at 25 C [bar]
p_25 = np.array([13.35,26.81,42.40,54.83,60.79])
# weight fraction at 25 C [w/w]
w_25 = np.array([0.0509,0.1088,0.1865,0.2711,0.3273])

# pressure at 60 C [bar]
p_60 = np.array([14.25,28.59,41.80,55.36,69.15,83.30,89.94])
# weight fraction at 60 C [w/w]
w_60 = np.array([0.0299,0.0617,0.0924,0.1253,0.1602,0.1974,0.2152])

fig1 = plt.figure()
ax = fig1.add_subplot(1,1,1)
ax.plot(p_25, w_25, 'o', color='b', label='25 C')
ax.plot(p_60, w_60, '^', color='r', label='60 C')
ax.set_xlabel('Pressure [bar]', fontsize=ax_fs)
ax.set_ylabel('weight fraction [w/w]', fontsize=ax_fs)
ax.set_title('CO2 in VORANOL 360', fontsize=t_fs)
ax.tick_params(axis='both', which='major', labelsize=tk_fs)
ax.set_xlim([0, 100])
ax.set_ylim([0, 0.3])
plt.legend(loc='best', fontsize=l_fs)
plt.savefig('Figures/solubility_co2_v360.pdf', bbox_inches="tight")
```

## Compare PPG and Voranol 2110b

Since they have essentially the same chemical structure and both have measurements at 25 C.

```python
# pressures for fit
p_fit = np.linspace(10,70,100)

# pressure at 25 C [bar]
p_25_2110b = np.array([13.66,27.97,41.80,55.64,61.65])
# weight fraction at 25 C [w/w]
w_25_2110b = np.array([0.0372,0.0821,0.1351,0.1993,0.2336])
# linear fit for visualization
a,b = np.polyfit(p_25_2110b, w_25_2110b, 1)
w_fit_2110b = a*p_fit + b

# pressure at 25 C [bar]
p_25_ppg = np.array([20,40,60])
# weight fraction at 25 C [w/w]
w_25_ppg = np.array([0.05, 0.11, 0.184])
# linear fit for visualization
a,b = np.polyfit(p_25_ppg, w_25_ppg, 1)
w_fit_ppg = a*p_fit + b

fig4 = plt.figure()
ax = fig4.add_subplot(1,1,1)
ax.plot(p_25_2110b, w_25_2110b, 'o', color='b', label='Voranol 2110b')
ax.plot(p_fit, w_fit_2110b, '--', color='k', label='Voranol 2110b - fit')
ax.plot(p_25_ppg, w_25_ppg, '^', color='b', label='PPG-2700')
ax.plot(p_fit, w_fit_ppg, '-.', color='k', label='PPG-2700 - fit')
ax.set_xlabel('Pressure [bar]', fontsize=ax_fs)
ax.set_ylabel('weight fraction [w/w]', fontsize=ax_fs)
ax.set_title('CO2 vs. P in PPG vs. Voranol 2110b', fontsize=t_fs)
ax.tick_params(axis='both', which='major', labelsize=tk_fs)
ax.set_xlim([0, 100])
ax.set_ylim([0, 0.3])
plt.legend(loc='best', fontsize=l_fs)
plt.savefig('Figures/solubility_co2_ppg_vs_v2110b.pdf', bbox_inches="tight")
```

## Compare PPG and Voranol 360

Voranol 360 is a mixture of PPG and propoxylated sorbitol, but couldn't hurt to show another comparison.

```python
# pressures for fit
p_fit = np.linspace(10,70,100)

# pressure at 25 C [bar]
p_25_360 = np.array([13.35,26.81,42.40,54.83,60.79])
# weight fraction at 25 C [w/w]
w_25_360 = np.array([0.0509,0.1088,0.1865,0.2711,0.3273])
# linear fit for visualization
a,b = np.polyfit(p_25_360, w_25_360, 1)
w_fit_360 = a*p_fit + b

# pressure at 25 C [bar]
p_25_ppg = np.array([20,40,60])
# weight fraction at 25 C [w/w]
w_25_ppg = np.array([0.05, 0.11, 0.184])
# linear fit for visualization
a,b = np.polyfit(p_25_ppg, w_25_ppg, 1)
w_fit_ppg = a*p_fit + b

fig4 = plt.figure()
ax = fig4.add_subplot(1,1,1)
ax.plot(p_25_360, w_25_360, 'o', color='b', label='Voranol 360')
ax.plot(p_fit, w_fit_360, '--', color='k', label='Voranol 360 - fit')
ax.plot(p_25_ppg, w_25_ppg, '^', color='b', label='PPG-2700')
ax.plot(p_fit, w_fit_ppg, '-.', color='k', label='PPG-2700 - fit')
ax.set_xlabel('Pressure [bar]', fontsize=ax_fs)
ax.set_ylabel('weight fraction [w/w]', fontsize=ax_fs)
ax.set_title('CO2 vs. P in PPG vs. Voranol 360', fontsize=t_fs)
ax.tick_params(axis='both', which='major', labelsize=tk_fs)
ax.set_xlim([0, 100])
ax.set_ylim([0, 0.3])
plt.legend(loc='best', fontsize=l_fs)
plt.savefig('Figures/solubility_co2_ppg_vs_v360.pdf', bbox_inches="tight")
```
