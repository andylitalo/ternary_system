# Rate of Sublimation of Dry Ice in Room Temperature Air

```python
import numpy as np
import matplotlib.pyplot as plt

# save parameters
savePlot = True
saveFolder = '..\\FIGURES\\'
saveName = 'dry_ice_m_vs_t'
ext = '.pdf'
# plot parameters
# font sizes
lFS = 14 # legend
axFS = 18 # axis labels
tFS = 20 # title
tkFS = 14 # tick labels
# plot sizes
LW = 2 # line width
MS = 8 # marker size
```

The data from the experiment, which was performed on December 4, 2018, and is recorded on p. 154 of Andy Ylitalo's lab notebook (1), are given below.

```python
# Time [s]
t = np.arange(0,240,10)
# mass [g]
m = np.array([10.258, 10.219, 10.130, 10.085, 10.041, 9.998, 9.954, 9.912, 9.871, 9.830, 9.790, 
             9.748, 9.707, 9.669, 9.629, 9.589, 9.551, 9.512, 9.469, 9.433, 9.394, 9.358, 9.320, 9.278])

# Linear regression
a, b = np.polyfit(t, m, 1)
t_reg = np.linspace(np.min(t),np.max(t))
m_reg = a*t_reg + b
# plot
fig = plt.figure(1)
plot = fig.add_subplot(111)
plot.plot(t, m, '^', label='data', markersize=MS)
plot.plot(t_reg, m_reg, 'r--', label=str(round(a,4)) + '*t + ' + str(round(b,4)), linewidth=LW)
plot.set_xlabel('time [s]', fontsize=axFS)
plot.set_ylabel('mass [g]', fontsize=axFS)
plot.set_title('Mass of Dry Ice over Time', fontsize=tFS)
plot.legend(loc='best', fontsize=lFS)
plot.tick_params(axis='both', which='major', labelsize=tkFS)

# save plot
if savePlot:
    fig.tight_layout()
    fig.savefig(saveFolder + saveName + ext, bbox='tight')
```
