# Plot Pressure Transducer Data

This file loads .csv files of data from a pressure transducer logged by an Arduino and plots them.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_p_vs_t(file_name, m, units='PSI', semilogx=False, x_lim=[]):
    """
    This function loads data for pressure vs. time in .csv format (pressure in first column,
    time in second) and plots.
    """
    # load data from file
    data = np.genfromtxt(file_name, delimiter=',')
    data.shape
    # parse data into pressure [units given] and time [seconds]
    p = data[:,0]
    t = data[:,1]

    # PLOT
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    if semilogx:
        ax.semilogx(t, p, 'o')
    else:
        ax.plot(t, p, 'o')
    ax.set_xlabel('time [s]')
    ax.set_ylabel('pressure [' + units + ']')
    ax.set_title('Pressure vs. Time with %.2f g Dry Ice' % m)
    if len(x_lim) == 2:
        ax.set_xlim(x_lim)
```

For this trial, I put in 6.06 g of dry ice, which should have resulted in 213 PSI according to the ideal gas law or 193 PSI according to the CO2 EoS (calculated using http://www.peacesoftware.de/einigewerte/co2_e.html).

```python
plot_p_vs_t('../DATA/co2_p_in_parr/20190206_p_psi_t_s.csv', m=6.06, x_lim=[0, 150*6.06])
```

Interesting, looks like there were two regimes of pressure increasing, a slower at first and a faster afterwards. Looks like the effect of temperature on the gas pressure over time was negligible.
