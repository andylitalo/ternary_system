We use the data from Di Caprio et al. 2016, *Fluid Phase Equilibria*, to produce a fit of mutual diffusivity of carbon dioxide in a blend of polyether polyols.

```python
import numpy as np
import matplotlib.pyplot as plt

# Functions
def lsq_err(x, y, y_fit, sigma_y, log=False):
    """
    Returns error of linear regression scaled by uncertainties sigma_y.
    """
    if log:
        residuals = ((np.log(y) - np.log(y_fit))/(sigma_y/y))**2
    else:
        residuals = ((y - y_fit)/sigma_y)**2
        
    return np.sum(residuals)
```

Next we put in the data.

```python
# mass fraction of co2 in polyol (g/g)
w = np.array([0.0064, 0.0132, 0.0216, 0.0235, 0.03147, 0.0374, 0.0414, 0.051, 0.052, 0.061, 0.066,
              0.071, 0.080, 0.094, 0.109, 0.123, 0.135])
# mutual diffusivity (cm^2/s)
D = np.array([6e-7, 1.0e-6, 1.1e-6, 8e-7, 1.4e-6, 1.5e-6, 1.6e-6, 1.2e-6, 1.3e-6, 3.8e-6, 1.7e-6,
              1.9e-6, 1.7e-6, 3.3e-6, 5.7e-6, 2.8e-6, 1.21e-5])
# uncertainty in mutual diffusivity (cm^2/s)
sigma_D = 5e-7*np.ones([len(D)])
```

Last we perform various fits and plot them
